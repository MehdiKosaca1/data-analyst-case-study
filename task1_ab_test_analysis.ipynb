## In-App Purchase (IAP) Revenue

```python
daily_purchase_ratio_A = 0.0305
daily_purchase_ratio_B = 0.0315
ad_impression_per_DAU_A = 2.3
ad_impression_per_DAU_B = 1.6
eCPM_A = 9.8
eCPM_B = 10.8
avg_purchase_value = 1
```

```python
from scipy.interpolate import interp1d

# --- Parametreler ---
installs_per_day = 20000
avg_purchase_price = 5.0  # Her satın alma ortalama 5 dolar
purchase_ratio_A = 0.0305
purchase_ratio_B = 0.0315

days_A = [0, 1, 3, 7, 14]
retention_A = [100, 53, 27, 17, 6]

days_B = [0, 1, 3, 7, 14]
retention_B = [100, 48, 25, 19, 9]

# --- Interpolasyon ---
f_linear_A = interp1d(days_A, retention_A, kind='linear', fill_value="extrapolate")
f_linear_B = interp1d(days_B, retention_B, kind='linear', fill_value="extrapolate")

# --- 15 Günlük Kümülatif IAP Geliri ---
total_iap_A = 0
cum_iap_A = []
total_iap_B = 0
cum_iap_B = []

for day in range(1, 16):
    # Variant A
    dau_A = sum((f_linear_A(age) / 100) * installs_per_day for age in range(0, day))
    daily_iap_A = dau_A * purchase_ratio_A * avg_purchase_price
    total_iap_A += daily_iap_A
    cum_iap_A.append(total_iap_A)
    
    # Variant B
    dau_B = sum((f_linear_B(age) / 100) * installs_per_day for age in range(0, day))
    daily_iap_B = dau_B * purchase_ratio_B * avg_purchase_price
    total_iap_B += daily_iap_B
    cum_iap_B.append(total_iap_B)

# --- Sonuç ---
print(f"Total IAP Revenue Variant A: ${total_iap_A:,.2f}")
print(f"Total IAP Revenue Variant B: ${total_iap_B:,.2f}")
diff = total_iap_A - total_iap_B
print(f"DIFFERENCE: ${abs(diff):,.2f} ({'Variant A is more profitable' if diff > 0 else 'Variant B is more profitable'})")
```

```python
# --- Grafik ---
plt.figure(figsize=(12,6), dpi=100)
x_axis = range(1, 16)

plt.plot(x_axis, cum_iap_A, color='#2563EB', marker='o', linewidth=3, label='Variant A')
plt.fill_between(x_axis, cum_iap_A, color='#2563EB', alpha=0.1)

plt.plot(x_axis, cum_iap_B, color='#DC2626', marker='s', linewidth=3, label='Variant B')
plt.fill_between(x_axis, cum_iap_B, color='#DC2626', alpha=0.1)

plt.title("15-Day Cumulative IAP Revenue Comparison", fontsize=14, fontweight='bold')
plt.xlabel("Day", fontsize=12)
plt.ylabel("Cumulative IAP Revenue ($)", fontsize=12)
plt.xticks(x_axis)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12, loc="upper left")
plt.show()

# --- Toplam Gelir ve Kazanan ---
print(f"Total IAP Revenue Variant A: ${total_iap_A:,.2f}")
print(f"Total IAP Revenue Variant B: ${total_iap_B:,.2f}")
diff = total_iap_A - total_iap_B
print(f"DIFFERENCE: ${abs(diff):,.2f} ({'Variant A is more profitable' if diff > 0 else 'Variant B is more profitable'})")
```

```python
# --- Parametreler ---
installs_old = 20000
installs_old_after20 = 12000
installs_new_after20 = 8000
avg_purchase_price = 5.0

# IAP
purchase_ratio_A = 0.0305
purchase_ratio_B = 0.0315

# Retention eski kullanıcılar
days_A = np.array([0, 1, 3, 7, 14])
retention_A = np.array([100, 53, 27, 17, 6])
days_B = np.array([0, 1, 3, 7, 14])
retention_B = np.array([100, 48, 25, 19, 9])

# Reklam
ecpm_A = 9.80
ad_imp_A = 2.3
ecpm_B = 10.80
ad_imp_B = 1.6
```

```python
# --- Exponential retention yeni kullanıcılar ---
def retention_new_A(x):
    return 0.58 * np.exp(-0.12 * (x-1)) * 100  # yüzdeye çevir

def retention_new_B(x):
    return 0.52 * np.exp(-0.10 * (x-1)) * 100  # yüzdeye çevir

# --- 30 Günlük Kümülatif Gelir (IAP + Ad) ---
total_rev_A = 0
cum_rev_A = []
total_rev_B = 0
cum_rev_B = []
```

```python
for day in range(1, 31):
    # --- Eski kullanıcılar DAU ---
    if day < 20:
        dau_A_old = sum((f_linear_A(age)/100) * installs_old for age in range(0, day))
        dau_B_old = sum((f_linear_B(age)/100) * installs_old for age in range(0, day))
    else:
        dau_A_old = sum((f_linear_A(age)/100) * installs_old_after20 for age in range(0, day))
        dau_B_old = sum((f_linear_B(age)/100) * installs_old_after20 for age in range(0, day))
    
    # --- Yeni kullanıcılar DAU (Day 20+) ---
    if day >= 20:
        dau_A_new = sum((retention_new_A(age)/100) * installs_new_after20 for age in range(0, day-19))
        dau_B_new = sum((retention_new_B(age)/100) * installs_new_after20 for age in range(0, day-19))
    else:
        dau_A_new = 0
        dau_B_new = 0
    
    # --- Günlük toplam DAU ---
    dau_A_total = dau_A_old + dau_A_new
    dau_B_total = dau_B_old + dau_B_new
    
    # --- Günlük toplam gelir ---
    daily_total_A = dau_A_total * (purchase_ratio_A * avg_purchase_price + ad_imp_A * ecpm_A / 1000)
    daily_total_B = dau_B_total * (purchase_ratio_B * avg_purchase_price + ad_imp_B * ecpm_B / 1000)
    
    # --- Kümülatif toplam gelir ---
    total_rev_A += daily_total_A
    total_rev_B += daily_total_B
    
    cum_rev_A.append(total_rev_A)
    cum_rev_B.append(total_rev_B)
```

```python
# --- Grafik ---
plt.figure(figsize=(12,6), dpi=100)
x_axis = range(1, 31)

plt.plot(x_axis, cum_rev_A, color='#2563EB', marker='o', linewidth=3, label='Variant A')
plt.fill_between(x_axis, cum_rev_A, color='#2563EB', alpha=0.1)

plt.plot(x_axis, cum_rev_B, color='#DC2626', marker='s', linewidth=3, label='Variant B')
plt.fill_between(x_axis, cum_rev_B, color='#DC2626', alpha=0.1)

plt.title("30-Day Cumulative Revenue with New User Source (IAP + Ad)", fontsize=14, fontweight='bold')
plt.xlabel("Day", fontsize=12)
plt.ylabel("Cumulative Revenue ($)", fontsize=12)
plt.xticks(range(1,31,2))
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12, loc="upper left")
plt.show()

# --- Sonuç ---
print(f"Total Revenue Variant A (Day 30 with New Source): ${total_rev_A:,.2f}")
print(f"Total Revenue Variant B (Day 30 with New Source): ${total_rev_B:,.2f}")
diff = total_rev_A - total_rev_B
print(f"DIFFERENCE: ${abs(diff):,.2f} ({'Variant A is more profitable' if diff > 0 else 'Variant B is more profitable'})")
```
